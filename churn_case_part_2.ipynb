# Analysis of Lifetime Value (LTV)

#####  Lifetime value (LTV) is the total worth to a business of a customer over the while period of their relationship. It's an important metric as it costs less to keep existings customers than to acquire new ones, so the idea of increasing the value of your existig customers is a great way to drive growth. 

```python
#importing library
import sqlite3 as sql
import pandas as pd
import numpy as np
```

```python
# creating connectiion and cursor
conn = sql.connect('churn.db')
cur = conn.cursor()
```

```python
df = pd.read_sql("select * from cust_final", conn)
df
```

# 1. Those customers who unsubscribed to service, what is their average LTV and how long do they stay alive in the service?

```python
churn_df = pd.read_sql("select * from cust_final where churn = 'Yes';", conn)
churn_df
```

```python
# finding distribution of total charges
churn_df['TotalCharges'].describe()
```

##### As we can see average total charges is very for 20% customers

```python
# finding the 80th percentile
churn_df['TotalCharges'].quantile(0.8)
```

```python
#Divide the distribution into below and above 80
total_charges_below_80 = churn_df.query("TotalCharges<=2840.41")
total_charges_above_80 = churn_df.query("TotalCharges > 2840.41")
```

```python
# show distribution of below 80 percentile
total_charges_below_80.TotalCharges.describe()
```

```python
# show distribution of above 80 percentile
total_charges_above_80.TotalCharges.describe()
```

```python
# show distribution of tenure of below 80 percentile
total_charges_below_80.Tenure.describe()
```

```python
# show distribution of tenure of above 80 
total_charges_above_80.Tenure.describe()
```

## The average LTV of those who unsubscribed to the service is $711 and tenure is approx. 10 months while for top 20%, average LTV is $4811 and tenure is near 50 months

# 2. When customers were subscribed, what services did they use? i.e. find top performing, worst performing services

```python
### Phone service percentage 
# below 80 
total_charges_below_80.PhoneService.value_counts()/total_charges_below_80.shape[0]
```

```python
#above 80
total_charges_above_80.PhoneService.value_counts()/total_charges_above_80.shape[0]
```

##### Only 2% of top 20% customers were not using phone services while in below 80% arond 11% were not using phone services i.e if you are not subscribed to phone service you are likely to be in below 80%

```python
#below 80
total_charges_below_80_use_phone = total_charges_below_80.query("PhoneService == 'Yes'")
total_charges_below_80_use_phone.MultipleLines.value_counts()/total_charges_below_80_use_phone.shape[0]
```

```python
#above 80
total_charges_above_80_use_phone = total_charges_above_80.query("PhoneService == 'Yes'")
total_charges_above_80_use_phone.MultipleLines.value_counts()/total_charges_above_80_use_phone.shape[0]
```

#### For the top 20% of LTV who used the phone service 84% of them used multiple lines, which is 2 times as much as the proportion of people in the low 80% LTV that is around 40%. Having multiple lines is going to get you closer to be in the top 20% of the LTV.

```python
## INternet under 80
total_charges_below_80.InternetService.value_counts()/total_charges_below_80.InternetService.shape[0]
```

```python
#above 80
total_charges_above_80.InternetService.value_counts()/total_charges_above_80.InternetService.shape[0]
```

- In the high-LTV group, all the customers used the internet service and 90% of them used a fiber optic connection.
- On the other hand, among the low-LTV (80%) customers, 7% did not use the internet service and only 64% of them used a fiber optic connection. 
##### This suggests that having internet services, and specifically a fiber optic connection, can bring you closer to the high-LTV group.
